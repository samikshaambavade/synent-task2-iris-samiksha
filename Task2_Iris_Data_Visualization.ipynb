```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
```

```python
df.isnull().sum()
```

```python
df["species"].value_counts().plot(kind="bar")
plt.title("Count of Iris Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
Observation:
The Iris dataset contains 50 samples of each species (Setosa, Versicolor, and Virginica). The dataset is perfectly balanced.
    
```

```python
df["sepal_length"].plot(kind="hist", bins=10)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.show()
Observation:
Most sepal lengths lie between 5 cm and 6.5 cm, showing that medium-sized flowers are the most common.
```

```python
plt.scatter(df["sepal_length"], df["petal_length"])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Sepal Length vs Petal Length")
plt.show()
Observation:
Setosa forms a clearly separate cluster, while Versicolor and Virginica overlap slightly.
```

```python
plt.scatter(df["petal_length"], df["petal_width"])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length vs Petal Width")
plt.show()
Observation:
Petal length and petal width vary significantly among the three species, making them useful features for classification.
```

# Conclusion

The Iris dataset was successfully visualized using different charts.

A bar chart showed the equal distribution of the three species.

The histogram displayed the distribution of sepal length values.

The scatter plot revealed that Setosa is clearly separated from the other species.

The comparison graph showed that petal measurements are the most useful features for distinguishing flower species.

These visualizations help understand the dataset before applying machine learning models.
